```python
%load_ext dotenv
%dotenv
%matplotlib inline

from xcube_sh.cube import open_cube
from xcube_sh.config import CubeConfig

cube = open_cube(cube_config=CubeConfig(
    dataset_name="S2L2A",
    band_names=["B04", "B08"],
    tile_size=[512, 512],
    geometry=(10.00, 54.27, 10.30, 54.50),
    spatial_res=0.00018,  # = 10.038 meters in degree
    time_range=["2018-03-14", "2018-06-20"],
    time_period="5D",
))
ndvi = (cube.B08 - cube.B04) / (cube.B04 + cube.B08)
ndvi.attrs['long_name'] = 'Normalized Difference Vegetation Index'
ndvi.attrs['units'] = 'unitless'
ndvi.plot.imshow(col='time', col_wrap=4, vmin=-1, vmax=1,cmap='RdYlGn')
```
